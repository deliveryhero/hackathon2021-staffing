```python
env = dict(
    DATA_LAKE_BUCKET="production-eu-data-science-data-lake",
    S3_PROFILE="log-prd",
    SAMPLING_MAX_RIDER_POSSIBLE_ASSIGNMENTS="10000",  # actually se uses 75,
    MIP_GAP_THRESHOLD="0.3",
    OVERWRITE_EXISTING_RESULTS="true"
)
os.environ.update(env)

city_id = 218  # Borlange
bucket = S3(bucket=env["DATA_LAKE_BUCKET"])
key = "raw_data/assignments/2021-07-26/se/input_25033,25022_2021-07-21T03:05:54.459684+00:00.pkl"

(
    starting_points,
    demand,
    employees,
    unavailabilities,
    opening_times,
) = bucket.download_pickle(key)
```

```python
from staffing.assignments.mip.mip_data import (
    generate_demand_indices,
    generate_under_staffing_penalty,
    generate_over_staffing_penalty,
)
from staffing.assignments.generate_possibilities import generate_possibilities
from staffing.unassigned_shifts.generate_shifts import generate_shifts

# demand_index is a dictionary listing the demand indexed by (starting_point, timestamp) pairs.
demand_index, scaled_demand_index, _ = generate_demand_indices(demand)

# potential_rider_shifts is dictionary indexed by the rider id mapping to a list of all potential shifts for a rider.
potential_rider_shifts, _ = generate_possibilities(
    city_id, employees, unavailabilities, opening_times
)

# shifts is a set of all potential shifts ignoring rider constraints. This is needed to create shift blocks.
shifts = {
    s
    for sp in starting_points
    for o in opening_times
    if o.zone_id == sp.geo_zone_id
    for s in generate_shifts(o.starts_at, o.ends_at, sp)
}

over_staffing_penalty = {
    di: generate_over_staffing_penalty(
        di[0],
        di[1],
        sp.staffing_rule.overstaffing_penalty,
        scaled_demand_index,
    )
    for sp in starting_points
    for di in demand_index.keys()
    if di[0] == sp.id
}
under_staffing_penalty = {
    di: generate_under_staffing_penalty(
        di[0],
        di[1],
        sp.staffing_rule.understaffing_penalty,
        scaled_demand_index,
    )
    for sp in starting_points
    for di in demand_index.keys()
    if di[0] == sp.id
}
```
